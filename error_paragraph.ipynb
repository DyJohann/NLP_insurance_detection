```python
import os
import re
```

```python
# Python內建filter()函式 - 過濾list
# filter()把傳入的函式依次作用於每個元素，然後根據返回值是True還是False決定保留還是丟棄該元素
def not_empty(s):
    return s and s.strip()
```

```python
def catch_csv(): 
    # 抓取目錄所有資料
    file_list = list()
    for root, dirs, files in os.walk("raw_data/test_predict/"):
        file_list = files
        break

    file_path_list = list()
    file_name_list = list()
    for file in file_list:
        if file.split(".")[-1] == "csv":
            file_path_list.append("raw_data/test_predict/"+file)
            file_name_list.append(file.split(".")[0])
    return file_path_list, file_name_list
```

```python
def catch_fail_paragraph(file_path, file_name):
    index = 0
    error_sentense_index = list()

    # 紀錄一段文字的label與各label機率
    text = ""
    true_label = ""
    predict_label = ""
    o_label = ""
    BHeading_label = ""
    IHeading_label = ""
    Bbenefit_label = ""
    Ibenefit_label = ""
    Bparameter_label = ""
    Iparameter_label = ""
    Bitem_label = ""
    Iitem_label = ""
    Brate_label = ""
    Irate_label = ""
    Bcurrency_label = ""
    Icurrency_label = ""

    # 儲存全部段落
    sentense_index = list()
    text_sentense = list()
    true_label_sentense = list()
    predict_label_sentense = list()
    o_label_sentense = list()
    BHeading_label_sentense = list()
    IHeading_label_sentense = list()
    Bbenefit_label_sentense = list()
    Ibenefit_label_sentense = list()
    Bparameter_label_sentense = list()
    Iparameter_label_sentense = list()
    Bitem_label_sentense = list()
    Iitem_label_sentense = list()
    Brate_label_sentense = list()
    Irate_label_sentense = list()
    Bcurrency_label_sentense = list()
    Icurrency_label_sentense = list()

    # 讀取CSV檔案
    with open(file_path, mode="r", encoding="utf-8-sig") as file:
        for line in file: #一行一行的讀取
            #print(line)
            line = re.split(",| |\n", line)
            line = list(filter(not_empty, line))

            if len(line) == 16:
                if line[0] != "text":
                    text += (line[0] + ",")
                    true_label += (line[1] + ",")
                    predict_label += (line[2] + ",")
                    o_label += (line[3] + ",")
                    BHeading_label += (line[4] + ",")
                    IHeading_label += (line[5] + ",")
                    Bbenefit_label += (line[6] + ",")
                    Ibenefit_label += (line[7] + ",")
                    Bparameter_label += (line[8] + ",")
                    Iparameter_label += (line[9] + ",")
                    Bitem_label += (line[10] + ",")
                    Iitem_label += (line[11] + ",")
                    Brate_label += (line[12] + ",")
                    Irate_label += (line[13] + ",")
                    Bcurrency_label += (line[14] + ",")
                    Icurrency_label += (line[15] + ",")
                    if (line[1] != line[2]) and (index not in error_sentense_index):
                        error_sentense_index.append(index)
            else:
                #print(text)
                text_sentense.append(text)
                true_label_sentense.append(true_label)
                predict_label_sentense.append(predict_label)
                o_label_sentense.append(o_label)
                BHeading_label_sentense.append(BHeading_label)
                IHeading_label_sentense.append(IHeading_label)
                Bbenefit_label_sentense.append(Bbenefit_label)
                Ibenefit_label_sentense.append(Ibenefit_label)
                Bparameter_label_sentense.append(Bparameter_label)
                Iparameter_label_sentense.append(Iparameter_label)
                Bitem_label_sentense.append(Bitem_label)
                Iitem_label_sentense.append(Iitem_label)
                Brate_label_sentense.append(Brate_label)
                Irate_label_sentense.append(Irate_label)
                Bcurrency_label_sentense.append(Bcurrency_label)
                Icurrency_label_sentense.append(Icurrency_label)

                # 清空文字
                text = ""
                true_label = ""
                predict_label = ""
                o_label = ""
                BHeading_label = ""
                IHeading_label = ""
                Bbenefit_label = ""
                Ibenefit_label = ""
                Bparameter_label = ""
                Iparameter_label = ""
                Bitem_label = ""
                Iitem_label = ""
                Brate_label = ""
                Irate_label = ""
                Bcurrency_label = ""
                Icurrency_label = ""
                sentense_index.append(index)
                index += 1
    
    # 寫入CSV檔案
    row_index = 1
    with open("raw_data/test_predict/analysis_" + file_name + ".csv", mode="w", encoding="utf-8-sig") as file:
        file.write(f'總段落數：,{index+1}\n')
        file.write(f'分辨錯誤段落數：,{len(error_sentense_index)}\n')
        file.write(f'錯誤率：,%.4f\n\n' % (len(error_sentense_index)/(index+1)))
        for index in error_sentense_index:
            file.write(str(row_index) + '\n')
            file.write('text,' + text_sentense[index] + '\n')
            file.write('true,' + true_label_sentense[index] + '\n')
            file.write('predict,' + predict_label_sentense[index] + '\n')
            file.write('o,' + o_label_sentense[index] + '\n')
            file.write('B-Heading,' + BHeading_label_sentense[index] + '\n')
            file.write('I-Heading,' + IHeading_label_sentense[index] + '\n')
            file.write('B-benefit,' + Bbenefit_label_sentense[index] + '\n')
            file.write('I-benefit,' + Ibenefit_label_sentense[index] + '\n')
            file.write('B-parameter,' + Bparameter_label_sentense[index] + '\n')
            file.write('I-parameter,' + Iparameter_label_sentense[index] + '\n')
            file.write('B-item,' + Bitem_label_sentense[index] + '\n')
            file.write('I-item,' + Iitem_label_sentense[index] + '\n')
            file.write('B-rate,' + Brate_label_sentense[index] + '\n')
            file.write('I-rate,' + Irate_label_sentense[index] + '\n')
            file.write('B-currency,' + Bcurrency_label_sentense[index] + '\n')
            file.write('I-currency,' + Icurrency_label_sentense[index] + '\n')
            file.write('\n')
            row_index += 1
    
    print("轉換完成：analysis_" + file_name + ".csv\n")
```
